```python
died     = titanic.loc[titanic['survived']==0, :]['pclass'].value_counts()
died
```

```python
survived = titanic.loc[titanic['survived']==1, :]['pclass'].value_counts()
died     = titanic.loc[titanic['survived']==0, :]['pclass'].value_counts()
df_plot  = pd.DataFrame([survived,died])
df_plot.index=['survived','died']

# Plot
df_plot.plot(kind='bar',stacked=True,colormap='Spectral', title='Stacked Bar plot');
```

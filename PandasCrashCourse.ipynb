# Lamda Function

```python
x = lambda a,b : a*b
y = x(4,5)
y
```

```python
def append_To(element, to=[]):
    to.append(element)
    return to

my_list = append_To(12)
print(my_list)
print('\n')
my_second_list = append_To(42)
print(my_second_list)
print('\n')
my_third_list = append_To(84,my_list.copy())
print(my_list)
print(my_third_list)
```
